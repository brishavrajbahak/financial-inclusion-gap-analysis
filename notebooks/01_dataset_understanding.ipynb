## 9. Day 2 Summary

### What the dataset contains

The dataset contains Global Findex 2025 financial inclusion indicators by country, year, region, income group, and demographic group. It includes indicator related to account owneership, financial institution account, mobile accounts, borrowing, saving, digital payments, and account inactivity.

### Dataset size
- Rows: 8577
- Columns: 438
- Years avaiable: 2011, 2014, 2017, 2021, 2024
- Nepal country code: NPL

### Likely unit of analysis

Each row appears to represent an aggregrated country-year-population-segment observation.

The orginal Global Findex data is based on individual survery responses, but this csv does not show individual survey respondents. Instead, it provides calculated indicators for countries and groups such as all adults, gender groups, income groups, and other population segments.

So, in this project, the unit of analysis should be treated as country/segment-level financial iunclusion analysis.

A practical row meaning is :

`country + year + group + group2`
